```python
# Importing sample data and libraries

file="1-100032-A-0.wav"

import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
%matplotlib inline


plt.figure(figsize=(14,5))
ipd.Audio(file)
```

```python
# Librosa
data,sample_rate=librosa.load(file)
librosa.display.waveshow(data,sr=sample_rate)

sample_rate
data.shape
```

```python
# Scipy
from scipy.io import wavfile as wav
wave_sample_rate, wave_audio=wav.read(file)

wave_sample_rate
#wave_audio

# Librosa normalizes values and has default sample rate as 22050.
# Librosa converts stereo into mono by combining two channels.
# Mono - Ex. Audio reaches both earbuds and Stereo - Some audio reaches both earbuds, or split
```
